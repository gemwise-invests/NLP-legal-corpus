```python
from sklearn.naive_bayes import MultinomialNB # 朴素贝叶斯
MultinomialNB = MultinomialNB()
cross_print_info(MultinomialNB, unigram, y, cv=10)
```

```python
from sklearn.naive_bayes import BernoulliNB # 朴素贝叶斯
BernoulliNB = BernoulliNB()
cross_print_info(BernoulliNB, unigram, y, cv=10)
```

```python
from sklearn.externals import joblib
Bernoulli= BernoulliNB.fit(unigram, y)
#joblib.dump(Bernoulli, '../classifier/Bernoulli.pkl') 
```

```python
from sklearn.ensemble import AdaBoostClassifier 
from sklearn.tree import DecisionTreeClassifier
reg = AdaBoostClassifier(DecisionTreeClassifier())
cross_print_info(reg, unigram, y, cv=10) 
```

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.model_selection import ShuffleSplit # Bootstrap抽样验证
ss = ShuffleSplit(n_splits = 20, test_size = 0.3, random_state=0)
rf = RandomForestClassifier(n_estimators=50)

cross_print_info(rf, unigram, y, cv = 10) # 随机森林
cross_print_info(rf, unigram, y, cv = ss)
```

```python
from sklearn.externals import joblib
rffit = rf.fit(unigram, y)
#joblib.dump(rffit, '../classifier/rffit.pkl') 
```

10 折交叉验证准确率为 0.766
Bootstrap抽样准确率为 0.772

```python
from xgboost import XGBClassifier # XGBoost
model = XGBClassifier(n_estimators = 100 ,colsample_bytree = 0.6, 
                      max_depth = 2, min_child_weight = 2, 
                      subsample = 0.8)
cross_print_info(model, unigram, y, cv = 10)
cross_print_info(model, unigram, y, cv = ss)
```

0.747 0.762

```python
from sklearn.externals import joblib
xgb = model.fit(unigram, y)
#joblib.dump(xgb, '../classifier/xgb.pkl') 
#clf = joblib.load('../classifier/xgb.pkl') 
```

10 折交叉验证准确率为 0.756
Bootstrap抽样准确率为 0.758

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
clf = model.fit(unigram, y) # XGBoost
y_predict = clf.predict(unigram)

print('训练准确率为',round(accuracy_score(y, y_predict),3))
```
